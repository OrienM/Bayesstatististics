## STEP2：查看需要反向计分的列  

1. 读取 **"EverythingIsEffort"** 列的数据  
2. 对这一列中的数值进行重新编码（如：将4重新编码为0，将3重新编码为1，将2重新编码为2，将1重新编码为3，将0重新编码为4） 并生成新的一列 **"EverythingIsEffort_recode"**  

此题可能的方法可参考：https://www.sfu.ca/~mjbrydon/tutorials/BAinPy/05_recode.html

```python
recode_mapping = {4:0, 3:1, 2:2, 1:3, 0:4}
df_dem["EverythingIsEffort_recode"] = df_dem["EverythingIsEffort"].replace(recode_mapping)
```

## STEP3: 计算几列分数的总和  
1. 提取数据框中的这几列： **"Nervous","Hopeless","RestlessFidgety","Depressed","EverythingIsEffort_recode","Worthless"**  
2. 计算这几列分数的总和，将结果储存在新的一列中  
3.  生成一个新的数据集**df_mental_sum**，其中一列是被试编号，命名为 **"worker_id"** ，另一列是分数总和，命名为 **"sum"**  
4.  **选取** "sum"列的最后10个值，将结果存在**a2**中

```python
df_dem['sum'] = df_dem[["Nervous", "Hopeless", "RestlessFidgety", "Depressed", "EverythingIsEffort_recode", "Worthless"]].sum(axis=1)
```

```python
df_mental_sum = df_dem[["worker_id","sum"]]
a2 = df_mental_sum["sum"].iloc[-10:]
```
