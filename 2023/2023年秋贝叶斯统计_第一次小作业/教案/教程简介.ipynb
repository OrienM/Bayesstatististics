**🔺在所有教程学习完毕后，务必要前往闯关题部分！**  

📍在学习与练习结束之后，你大致可以学会：  

>* **生成模拟数据**  
>* **完成基本的数据清洗**  
>* **进行基本的数据可视化**  

*教程中的例子以心理学实验中常见的问卷数据与反应时数据为主*
