```python
import seaborn as sns
import pandas as pd
```

```python
# 导入seaborn库中带有的数据集
penguins = pd.read_csv("/home/mw/input/37953795/penguins.csv")

#查看前五行数据
penguins.head()
```

## 箱图

```python
# 指定x轴与y轴代表的变量
sns.boxplot(data=penguins, 
            x="species", 
            y="flipper_length_mm")

# 去掉右边和上面的边框，使坐标轴更符合APA要求
sns.despine()
```

## 条形图

```python
# 指定x轴与y轴代表的变量，使用hue来指定第二个分类标准：性别
sns.barplot(data = penguins, 
            x = "species", 
            y = "flipper_length_mm", 
            hue="sex",
            errorbar="sd")
sns.despine()
```

## 散点图

```python
# 指定x轴与y轴代表的变量
sns.scatterplot(data = penguins, 
                x = "bill_length_mm", 
                y = "bill_depth_mm")
sns.despine()
```

```python
# 增加拟合的回归直线
sns.regplot(data = penguins, 
                x = "bill_length_mm", 
                y = "bill_depth_mm")
sns.despine()
```

## 直方图/密度图

```python
sns.displot(penguins, x = 'flipper_length_mm')
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde")
```

```python
# 通过kde将两种图结合在一起
sns.histplot(data=penguins, x="flipper_length_mm", kde=True)
sns.despine()
```

```python
# 通过hue来指定图示中新的类别变量
sns.displot(penguins, x="flipper_length_mm", hue="species")
```

# 多图拼接

```python
sns.displot(penguins, x="flipper_length_mm", col="species")
```

```python
sns.displot(
    penguins, 
    x="flipper_length_mm", 
    col="species",                      #指定列变量(分类变量)
    row="sex",                          #指定行变量(分类变量)
    binwidth=3,                         #调整每一个矩形的宽度
    height=3,
    facet_kws=dict(margin_titles=True)) #将行变量的标题(即sex)绘制在最右侧
```
