### 模型假设的影响

**🤔 我们知道模型存在前提假设，如果这些模型的前提假设不成立，模型会变得多糟糕？**  

在lec9中，我们使用一个线性模型来定义压力分数与自我控制分数之间的关系，并且指定了该模型成立的一些前提假设“

$$  
\begin{align*}  
Y_i &= \beta_0 + \beta_1 X_i + \epsilon \;\;\;\;\;\;\;\;\epsilon \sim N(0,\sigma^2)\\  
&\Downarrow \\  
Y_i | \beta_0, \beta_1, \sigma &\stackrel{ind}{\sim} N\left(\mu_i, \sigma^2\right) \;\; \text{ with } \;\; \mu_i = \beta_0 + \beta_1X_i  .\\  
\end{align*}  
$$  

* 回归模型需满足如下假设：  

    1. 独立观测假设:每个观测值$Y_i$是相互独立的，即一个观测的值不受其他观测的影响  

    2. 线性关系假设: 预测值$\mu_i$和自变量$X_i$之间可以用线性关系来描述，即：$\mu_i = \beta_0 + \beta_1 X_i$  

    3. 方差同质性假设： 在任意自变量的取值下，观测值$Y_i$都会以$\mu_i$为中心，同样的标准差$\sigma$呈正态分布变化（$\sigma$ is consistent）  

![Image Name](https://cdn.kesci.com/upload/s3y4rmsym1.png?imageView2/0/w/960/h/960)



**1. 当假设1(独立观测假设)被违反时：**

* 在心理学的实验数据当中，很多时候观测值之间存在依赖关系(dependent)。
* 比如，反应时在单个被试内、或某种特定刺激类型内可能更同质：有的被试总是反应得比其他人更快，有的刺激总是能导致更快的反应。
* 这时，被试内/项目内反应时数据的相互关联会导致对于结果的错误估计。

* 我们在之后会学习使用层级贝叶斯模型来处理相互关联的数据。

![Image Name](https://cdn.kesci.com/upload/s3xt2u50rw.png?imageView2/0/w/960/h/960)

> source: Brown, V. A. (2021). An Introduction to Linear Mixed-Effects Modeling in R. Advances in Methods and Practices in Psychological Science, 4(1), 2515245920960351. https://doi.org/10.1177/2515245920960351

**2. 当假设2(线性关系假设)和假设3(方差同质性假设)被违反时：**

* 假设2(线性关系假设)违反的情况：在左图中，我们可以看到，Y和X之间的关系并非线性的

* 假设3(方差同质性假设)违反的情况：并且，随着X的增大，Y的变异性越来越大。

  * 这要导致的后果是，后验预测分布比实际观测值的分布差异很大(右图)

![Image Name](https://www.bayesrulesbook.com/bookdown_files/figure-html/nonlinear-1-ch10-1.png)



### 对模型进行修改

* 考虑到并不是所有数据都会满足这些假设，当这些假设被违反时，我们需要考虑模型的修改。

对于违反假设1 的情况，我们在之后会学习使用层级贝叶斯模型来处理相互关联的数据。

**对于违反假设2和3的情况，通常有两种处理方式**

**a. 使用不同的数据模型**

* 不假设实际值与观测值之间的关系是正态的  $Y_i \sim N(\mu_i, \sigma^2)$

* 之后，我们会学习使用其他回归模型来描述其数据关系，比如泊松回归、二项回归、负二项回归。

**b. 对数据进行变换**

> 如果数据模型并不是我们要担心的问题，我们可以对数据进行变换，仍然可以对变换后的数据可以使用正态模型：

* 对$Y$进行变换：$g(Y_i) | \beta_0, \beta_1, \sigma \stackrel{ind}{\sim}N(\mu_i, \sigma^2)$，$\mu_i = \beta_0 + \beta_1 X_i$

* 对$X$进行变换： $Y_i | \beta_0, \beta_1, \sigma \stackrel{ind}{\sim}N(\mu_i, \sigma^2)$，$\mu_i = \beta_0 + \beta_1 h(X_i)$

* 同时对$Y$和$X$进行变换：$g(Y_i) | \beta_0, \beta_1, \sigma \stackrel{ind}{\sim}N(\mu_i, \sigma^2)$， $\mu_i = \beta_0 + \beta_1 h(X_i)$

在刚刚的例子当中，我们对$Y$的取值做一个对数变换

$$\log(Y_i) | \beta_0, \beta_1, \sigma \stackrel{ind}{\sim}N(\mu_i, \sigma^2) \;\; \text{ with } \; \mu_i = \beta_0 + \beta_1 X_i$$

* 在变换之后，可以看到$\log(Y)$与$X$之间的关系仍然是线性的，且随着$X$的增大，$Y$的变异性仍然是一致的。

* 可以使用正态的线性模型来拟合$\log(Y)$的后验分布

## 模型评估

一般情况下，我们的贝叶斯模型不会是完全不公平的，或者错得太离谱的。 但除了这些问题，**更为重要的是，模型是否可以用来准确预测新数据 Y 的结果。**

> 如果说模型公平性和模型错误描述的是模型在质量上的优劣，那模型评估与比较就是在数量上衡量模型预测的准确性。

接下来我们讲介绍衡量后验预测的三种指标：
- 绝对误差的中位数，median absolute error (MAE)
- 后验预测区间，posterior prediction interval

### Median absolute error (MAE)

绝对误差的中位数，median absolute error (MAE)衡量了预测观测值$Y_i$与后验预测均值之间$Y_i'$的差异。

$$
\text{MAE} = median\left(|Y_i - \hat{Y}_i|\right)
$$

- 假设 Y1 Y2…， Yn表示n个观察结果。
- 每个 $Y_i$ 都有对应的后验预测模型，其均值为 $Y_i'$，其标准差为 $sd_i$。

结合我们之前的例子来理解绝对误差的中位数：
- 首先，我们以压力 $X_i=45$ 下的后验预测模型为例子。
- 之后，我们计算后验预测模型的均值和观测值之间的误差。
- 再次，我们计算所有观测变量下后验预测模型的均值和观测值之间的误差。
- 最后，我们计算所有误差的中位数。

**首先，我们考虑压力 $X_i=45$ 下的后验预测模型**

1. 和之前一样，我们首先读取数据 Data_Sum_HPP_Multi_Site_Share，选择"Tsinghua"站点的数据作为示例。
2. 之后，使用 pymc 对压力变量`stress` 和自我控制变量 `scontrol` 建立回归模型。
3. 在模型拟合后(略过模型诊断环境)，生成模型的后验预测。
4. 最后，我们计算压力 $X_i=45$条件下，后验预测的均值和观测值之间的误差。

```python
# 导入 pymc 模型包，和 arviz 等分析工具 
import pymc as pm
import arviz as az
import seaborn as sns
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd

# 忽略不必要的警告
import warnings
warnings.filterwarnings("ignore")
```

```python
# 通过 pd.read_csv 加载数据 Data_Sum_HPP_Multi_Site_Share.csv
df_raw = pd.read_csv('../lin/lec10/Data_Sum_HPP_Multi_Site_Share.csv')

# 筛选站点为"Tsinghua"的数据
df = df_raw[df_raw["Site"] == "Tsinghua"]

# 选择需要的两列
df = df[["stress","scontrol"]]

#设置索引
df["index"] = range(len(df))
df = df.set_index("index")
```

```python
# 对压力变量`stress` 和自我控制变量 `scontrol` 建立回归模型
with pm.Model() as linear_model:

    beta_0 = pm.Normal("beta_0", mu=70, sigma=4)          #定义beta_0          
    beta_1 = pm.Normal("beta_1", mu=0, sigma=0.8)         #定义beta_1
    sigma = pm.Exponential("sigma", 0.6)                  #定义sigma

    #x是自变量压力水平
    x = pm.MutableData("x",df.stress,dims="obs_id")       
    #定义mu，讲自变量与先验结合              
    mu = pm.Deterministic("mu", beta_0 + beta_1 * x, dims="obs_id")    
    #定义似然：预测值y符合N(mu, sigma)分布, observed 为传入实际数据y 自我控制水平
    likelihood = pm.Normal("y_est", mu=mu, sigma=sigma, observed=df.scontrol, dims="obs_id")   
```

```python
# 模型拟合 
#===========================
#     注意！！！以下代码可能需要运行1-2分钟左右
#===========================
with linear_model:
    trace = pm.sample(draws=5000,                   # 使用mcmc方法进行采样，draws为采样次数
                      tune=1000,                    # tune为调整采样策略的次数，可以决定这些结果是否要被保留
                      chains=4,                     # 链数
                      discard_tuned_samples= True,  # tune的结果将在采样结束后被丢弃
                      random_seed=84735)
```

接下来，我们生成压力 $X_i=45$条件下的后验预测模型，并计算后验预测的均值和观测值之间的误差

- 通过 `pm.set_data({"x": np.array([45])})` 设置模型生成x=40条件下的后验预测模型
- 由于自定义的数据可能为样本外数据，因此需要设置 `predictions=True`进行样本外预测。
- 样本外预测结果储存在 `ppc.predictions` 中。而 `ppc.predictions_constant_data`是我们设置的x=40的数据。

```python
# 生成模型的后验预测
with linear_model:

    pm.set_data({"x": np.array([45])})   # 注意：仅生成x=45条件下的后验预测模型

    ppc = pm.sample_posterior_predictive(
        trace, 
        var_names=["y_est"],
        predictions=True,
        random_seed=84735
    )
```

我们可以计算后验预测的均值:

- 我们计算在所有链和参数样本条件下的后验预测的均值
- 结果显示，在压力 $X_i=45$时，后验预测(自我控制得分)的均值为36.9

```python
ppc.predictions["y_est"].stack(sample=("chain", "draw")).mean()
```

我们可以进一步计算后验预测误差(prediction error)。
- $\text{prediction error} = Y-Y'$

$$
\begin{array}{lll}

\left[
\begin{array}{lll}

\hline
\beta_0^{(2)} & \beta_1^{(2)} & \sigma^{(2)} \\
\hline

$$

**计算所有误差的中位数(MAE)**

$$
{\mathsf{MAE}}={\mathsf{median}}|Y_{i}-Y_{i}^{\prime}|.
$$

- MAE 为所有预测误差Y-Y'的中位数。

最后，我们计算所有观测值与后验预测模型均值之间的误差。

- 首先，我们将之前计算后验预测误差(prediction error)的方法封装为函数，方便后续计算。
- 然后，使用循环方法，计算每种X调节下的预测误差。
- 最后，合并所有误差，并计算这些误差的中位数。

![Image Name](https://cdn-media-1.freecodecamp.org/images/MNskFmGPKuQfMLdmpkT-X7-8w2cJXulP3683)

资料来源：https://www.freecodecamp.org/news/machine-learning-mean-squared-error-regression-line-c7dde9a26b93/

```python
def prediction_error(X=45):
  # 提取对应X值下的观测值Y
  true_ys = df[df.stress == X]["scontrol"].values
  # 提取对应X值下的后验预测模型的均值
  pre_y_mean_i = pre_y_mean[pre_y_mean.stress == X]["scontrol_ppc_mean"].values
  # 计算预测误差
  pre_error = true_ys - pre_y_mean_i
  return pre_error

# 计算X为45时的预测误差
print("prediction error:\n", prediction_error(X=45))
```

```python
# 导入 median 函数，用于计算中位数
from statistics import median
# 定义空列表，用于合并所有预测误差
pre_errors = []


# 使用循环方法，计算每种X调节下的预测误差。
for x in x_stress:                      # x_stress 我们获得的X的32种可能的取值
  pre_errors.append(prediction_error(X=x))

# 最后，计算预测误差的中位数。
pre_errors = np.concatenate(pre_errors)  # 将预测误差合并为numpy array
MAE = median(pre_errors)
print(f"计算得到的MAE: {MAE:.2f}")
```

```python
# 可视化预测误差和MAE
az.plot_posterior(pre_errors, hdi_prob=0.95, point_estimate="median", ref_val=MAE)
```

对于 MAE 的解读。

- 绝对误差的中位数，median absolute error (MAE)衡量了预测观测值$Y_i$与后验预测均值之间$Y_i'$的差异。
- **MAE越小**表明后验模型的**预测越准确**。



### 后验预测区间，posterior prediction interval

MAE 是一个有效的指标，可以衡量模型对于样本数据的预测能力。

然而，MAE 仅考虑了后验预测模型的均值，而并不是其可信区间。

后验预测区间(posterior prediction interval)则进一步反应了观测数与后验预测可信之间的关系。

$$

* 在当前数据集中，我们可以计算 32 种X取值下对应的后验预测模型的均值。
  * 代入`20000`个参数对，每个自变量都拥有对应的20000个后验预测值
* 进而，我们可以计算出32个可信区间，并观察真实值落在可信区间的情况。


```python
# 获取x压力数据的32种可能的值
x_stress = df.stress.unique()

# 生成32种X取值下对应的后验预测模型
with linear_model:

    pm.set_data({"x": x_stress})   # 设置生成32种X条件下的后验预测模型

    ppc = pm.sample_posterior_predictive(
        trace, 
        var_names=["y_est"],
        predictions=True,
        extend_inferencedata=True,
        random_seed=84735
    )
```

我们将之前计算单个X对应下后验预测模型HDI的方式打包为函数：

- 首先，定义函数，计算对应X下的后验预测模型hdi范围
- 其次，判断观测值是否超过 hdi 边界
- 最后，记录并返回超过 hdi 边界的观测值的数量

```python
# 定义函数，计算对应X下的观测值超出后验预测模型hdi范围的数量
def count_outlier(i,hdi=0.95):
  # 提取第i个X的值
  X = ppc.predictions_constant_data.x.isel(obs_id=i).values
  # 提取第i个X值对应的后验预测值
  pre_ys = ppc.predictions["y_est"].stack(sample=("chain", "draw")).isel(obs_id=0).values
  # 提取对应X值下的观测值Y
  true_ys = df[df.stress == X]["scontrol"]
  
  # 计算对应后验预测的hdi的上下限
  lower,upper = az.hdi(pre_ys, hdi_prob=hdi)
  
  # 判断是否超过 hdi 边界
  outlier = (true_ys > upper) | (true_ys < lower)

  # 计算超出边界的观测值的数量
  num_outlier = sum(outlier)
  return num_outlier

# 计算X为45时的预测误差
i = 5
print(f"X为{x_stress[i]}下，超过后验预测范围观测值的数量:", count_outlier(i))
```

接着，我们可以通过循环的方式计算每个X对应下超过后验预测范围观测值的数量。

```python
# 定义所有超过后验预测范围观测值的数量
total_number = 0

# 使用循环方法，循环32种X调节下的预测误差。
for i in range(32):  
  total_number += count_outlier(i)

print(f"所有超过后验预测范围观测值的数量: {total_number}")
```

考虑到我们一共有 174 个观测值，其中有17个观测值超过95%的后验预测范围。

这表明，我们的预测准确率超过 90%，$(174-17)/174 = 0.902$。
