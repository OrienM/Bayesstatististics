艺术深深植根于人类的天性，并深刻地影响着人类的情感、思维和行为。每当我们欣赏精妙绝伦的艺术作品、历史悠久的人文名胜、鬼斧神工的自然景观时，都能获得与艺术交融汇通的精神上的满足，内心均会油然而生一种**敬畏之情**。  

**在个体面对相对于自身来说是巨大的并且超越了之前认识的自然和人造的物体，就可以引起敬畏的体验（Keltner & Haidt, 2003）。**  


![Image Name](https://cdn.kesci.com/upload/s69xd1u7hz.jpg?imageView2/0/w/640/h/640)  

在这种审美情趣的敬畏里，我们感受到到内心的平静和超然，而这与更健康的身体和更良好的社会功能息息相关(Stellar et al., 2017)。  

**研究发现，成年人的敬畏经历与更多的副交感神经系统（PNS）激活有关，PNS活动的增加（即RSA的增加）支持儿童和成年人平静的社会参与(Gordon et al., 2017)。**  

注：迷走神经控制呼吸性窦性心律不齐（RSA），即心率和呼吸的协同变化，这通常被认为是心脏健康的指标和情绪调节能力提升的预测因子 (Kok et al., 2013)

敬畏也可以**提升成年人的亲社会水平** (Keltner et al., 2014)。敬畏让人们感到自我的渺小、谦卑(Stellar et al., 2018)，并将注意力从自我转移到他人需求和社会世界上，从而导致人们做出更多的有益于他人的行为。  
**而在儿童身上我们也能看到类似的线索。**比如，5个月大的婴儿更容易对莫扎特的音乐微笑。  

而由于以往研究对积极情绪的忽视，并且假设敬畏这种复杂情绪更可能和个体后期发展阶段更同步，导致了**对童年期的敬畏研究不足**（Fredrickson., 2001)，值得注意的是，迄今为止还没有激发和测量儿童敬畏情绪的有效方法。

而在儿童的亲社会研究中发现，儿童帮助他人的内群体偏好在发展早期已经出现，比如，当与他人分享时，4-6岁的儿童对外群体接受者的慷慨分配少于内群体接受者；与来自另一所学校的孩子相比，3-8岁的孩子也更有可能与同学分享糖果。  
因此，本研究关注**艺术引起的敬畏对儿童对外群体的亲社会行为的影响及其独特的生理关联**。在两项预先登记的研究中(研究1:N = 159，研究2:N = 353)， 8 - 13岁的孩子观看了引起敬畏、喜悦或中性(对照)反应的电影片段。观看了这段令人敬畏的视频的孩子更有可能在亲社会任务中表现出更多的耐心(研究1)，并把他们的实验收入捐赠出去(研究1和2)用于帮助难民。他们还表现出增加的呼吸性窦性心律不齐，这是一个与社会参与有关的副交感神经系统激活的指标。我们讨论了通过重新想象儿童的环境来激发关键年龄阶段的敬畏，从而培养亲社会行为的意义。

## Poisson regression：  

1. 自变量：（emotion condition）快乐、敬畏  

2. 因变量：计数任务数量  

3. 数据关系：  
$$  
Y_i | \beta_0,\beta_1 \stackrel{ind}{\sim} Pois\left(\lambda_i\right) \;\; \text{ with } \;\; \lambda_i = e^{\beta_0 + \beta_1 X_{i1}}  
$$  

$$  
\begin{array}{lcrl}  
\text{priors:} & & \beta_{0}  &  \sim N\left(1, 1^2 \right)  \\  
               & & \beta_1  & \sim N\left(1, 1^2 \right)\\  
\end{array}  
$$  

### 泊松分布模型设定、先验选择

```python
with pm.Model(coords = {"obs_id": df.index})  as poisson_model:

   beta_0 = pm.Normal("beta_0",mu=0.5,sigma=1)         #定义beta_0          
   beta_1=pm.Normal("beta_1",mu=0.5,sigma=1)           #定义beta_1
   #自变量:情绪唤醒条件
   emotion = pm.MutableData("emotion",df.EmotionCondition_recoded)
   ##定义mu，自变量与先验结合      
   lam = pm.Deterministic("lam", 
                           pm.math.exp(beta_0 + beta_1*emotion), 
                           dims="obs_id")  
   #定义泊松分布的似然
   likelihood = pm.Poisson("y_est", mu=lam, observed=df.TotalItemCountAccurate, dims="obs_id")
```

```python
#使用 pm.sample_prior_predictive 函数从先验分布中生成预测样本
poisson_prior =pm.sample_prior_predictive(samples=50, 
                                          model=poisson_model,
                                          var_names=["lam"],
                                          random_seed=84735)
```

### 先验检验

```python
for i in range(50):
    plt.plot(poisson_prior.constant_data["emotion"],
             poisson_prior.prior["lam"].stack(sample=("chain", "draw"))[:, i], 
             c="grey",
             alpha=0.4)
# 设置x、y轴标题和总标题    
plt.xlabel("emotion", fontsize=12)
plt.ylabel("count", fontsize=12)
plt.suptitle("50 prior plausible models", fontsize=14)
sns.despine()
plt.show()
```

### MCMC评估

```python
with poisson_model:
    poisson_model_trace = pm.sample(
        draws=5000,
        tune=1000,
        chains=4,
        discard_tuned_samples=True,
        idata_kwargs={"log_likelihood": True},
        random_seed=84735
    )
```

```python
az.plot_trace(poisson_model_trace, var_names=["beta_0","beta_1"],
              figsize=(15,10))
plt.show()
```

```python
az.summary(poisson_model_trace, var_names=["beta_0","beta_1"])
```

### 后验预测评估

```python
for i in range(100):
    sns.lineplot(x = poisson_model_trace.constant_data["emotion"],
            y = poisson_model_trace.posterior["lam"].stack(sample=("chain", "draw"))[:,i], 
            c="grey",
            alpha=0.4)
#设置x、y轴标题和总标题    
plt.xlabel("emotion",
           fontsize=12)
plt.ylabel("count",
           fontsize=12)
plt.suptitle("100 posterior plausible models",
           fontsize=14)
sns.despine()
plt.show()
```

### 模型评估

```python
#从后验分布中生成后验预测样本
with poisson_model:
    poisson_model_ppc = pm.sample_posterior_predictive(poisson_model_trace, random_seed=84735)
```

### 负二项回归模型设定、先验选择

##  Negative Binomial model：  

1. 自变量：（emotion condition）快乐、敬畏  

2. 因变量：计数任务数量  

3. 数据关系：  
$$  
\begin{equation}  
\begin{array}{lcrl}  
\text{data:} & \hspace{.025in} & Y_i|\beta_0,\beta_1,r & \stackrel{ind}{\sim} \text{NegBin}\left(\mu_i, r \right) \;\; \text{ with } \;\; \mu_i= e^{\beta_0 + \beta_1 X_{i1}}\\  
\text{priors:} & & \beta_{0}  & \sim N\left(1, 1,1^2\right) \\  
 & & \beta_1  & \sim N\left(1, 1^2\right) \\  
& & r & \sim \text{Exp}(1)\\  
\end{array}  
\end{equation}  
$$

```python
with pm.Model(coords = {"obs_id": df.index}) as nb_model:

    beta_0 = pm.Normal("beta_0", mu=1,sigma=1)         #定义beta_0          
    beta_1=pm.Normal("beta_1", mu=1,sigma=1)           #定义beta_1
    alpha = pm.Exponential("alpha", 1)
    emotion = pm.MutableData("emotion",df.EmotionCondition_recoded)               
    lam = pm.Deterministic("lam", 
                           pm.math.exp(beta_0 + beta_1*emotion), 
                           dims="obs_id")    #定义mu，自变量与先验结合
    
    likelihood = pm.NegativeBinomial("y_est", mu=lam, alpha=alpha, observed=df.TotalItemCountAccurate)
```

```python
def count_outlier(i,hdi=0.94,ppc=nb_model_ppc):

  # 提取第i个X的值
  X = ppc.constant_data["emotion"][i].values
  # 提取第i个X值对应的后验预测值
  pre_ys = ppc.posterior_predictive["y_est"].stack(sample=("chain", "draw"))[i].values
  # 提取对应X值下的观测值Y
  true_ys = ppc.observed_data["y_est"][i].values
  
  # 计算对应后验预测的hdi的上下限
  lower,upper = az.hdi(pre_ys, hdi_prob=hdi)
  
  # 判断是否超过 hdi 边界
  outlier = (true_ys > upper) | (true_ys < lower)

  return outlier
```

```python
hdi_verify = []

# 使用循环，判断所有X对应的Y是否超出后验预测范围，并计数
for i in range(len(df)):  
  hdi_verify.append(count_outlier(i, hdi=0.94))
total_number = sum(hdi_verify)
print(f"所有超过后验预测范围HDI的数量: {total_number}")
```

## 逻辑回归-Bernoulli  

1. 自变量：（emotion condition）快乐、敬畏  

2. 因变量：捐赠行为  

3. 数据关系：  
$$  
\begin{array}{lcrl}  
\text{data:} & \hspace{.01in} & Y_i|\beta_0,\beta_1 & \stackrel{ind}{\sim} \text{Bern}(\pi_i) \;\; \text{ with } \;\; \pi_i = \frac{e^{\beta_0 + \beta_1 X_{i1}}}{1 + e^{\beta_0 + \beta_1 X_{i1}}} \\  
\text{priors:} & & \beta_{0}  &  \sim N\left(0.3, 0.4^2 \right)  \\  
               & & \beta_1  & \sim N\left(0.3, 0.4^2 \right)\\  
\end{array}  
$$

### 模型定义、先验选择

```python
with pm.Model() as log_model1:
    # 此处对coords的定义方式进行了更改，因为后续我们需要进行对新数据的预测
    # 因此将维度定义成可更改的
    log_model1.add_coord('obs_id',df.index, mutable=True)

    y = pm.MutableData('y', df.DonateTicket, dims = 'obs_id')

    beta_0 = pm.Normal("beta_0",mu=0.5,sigma=0.5)          #定义beta_0          
    beta_1 = pm.Normal("beta_1",mu=0.5,sigma=0.5)          #定义beta_1
    emotion = pm.MutableData("emotion",df.EmotionCondition_recoded)                                            
    mu = pm.Deterministic("mu", 
                           beta_0 + beta_1*emotion, 
                           dims="obs_id")    
    #相当于进行了如下计算 (1 / (1 + exp(-mu))
    pi = pm.Deterministic("pi", pm.math.invlogit(mu), dims="obs_id")
    #似然
    likelihood = pm.Bernoulli("y_est",p=pi, observed=y,dims="obs_id")
```

### 模型1：RSA的固定截距模型  
$$  
[ \text{RSA}_{\text{fixedintercept}} = \beta_0 + \beta_1 \times \text{Time} + \beta_2 \times \text{EmotionCondition} + \beta_3 \times \text{baseline_mean_RSA} + \beta_4 \times \text{practice_mean_RSA} + \beta_5 \times \text{Time} \times \text{EmotionCondition} + \epsilon ]  
$$

```python
# 创建固定截距RSA模型
with pm.Model() as RSA_fixedintercept_model:

    # 模型设定
    beta_0 = pm.Normal("beta_0", mu=0, sigma=1)  
    beta_time = pm.Normal("beta_time", mu=0, sigma=1)
    beta_emotion_2 = pm.Normal("beta_emotion_2", mu=0, sigma=1)
    beta_emotion_4 = pm.Normal("beta_emotion_4", mu=0, sigma=1)
    beta_baseline_RSA = pm.Normal("beta_baseline_RSA", mu=0, sigma=1)
    beta_practice_RSA = pm.Normal("beta_practice_RSA", mu=0, sigma=1)
    

    # 线性模型
    mu = pm.Deterministic("mu",
        beta_0 +
        beta_time * df_subset["Time"]+
        beta_emotion_2 * df_subset["Emotion_1"] +
        beta_emotion_4 * df_subset["Emotion_1"] * df_subset["Time"]+
        beta_baseline_RSA * df_subset["baseline_mean_RSA"] +
        beta_practice_RSA * df_subset["practice_mean_RSA"]
    )
    
    # 定义观测模型
    sigma = pm.HalfCauchy("sigma", beta=1)
    likelihood = pm.Normal("y_RSA", mu=mu, sigma=sigma, observed=df_subset["RSA"])

```

### 模型2：RSA的随机截距模型（被试随机截距）  
$$  
[ \text{RSA}_{\text{randomintercept}} = \beta_0 + \beta_1 \times \text{Time} + \beta_2 \times \text{EmotionCondition} + \beta_3 \times \text{baseline_mean_RSA} + \beta_4 \times \text{practice_mean_RSA} + \beta_5 \times \text{Time} \times \text{EmotionCondition} + \text{random_intercept} + \epsilon ]  
$$

```python
def run_RSA_randomintercept_model(non_centered=False):

    # 定义数据坐标，包括被试和观测索引
    coords = {"ppID": df_subset['ppID_new'].values,  # 使用 .values 属性将 Pandas Series 转换为 NumPy 数组
              "obs_total": df_subset['obs_total'].values}
      
    with pm.Model(coords=coords) as RSA_randomintercept_model:
        # 定义全局参数
        beta_0 = pm.Normal("beta_0", mu=5, sigma=1)
        beta_0_sigma = pm.Exponential("beta_0_sigma", 1)
        beta_1 = pm.Normal("beta_1", mu=0, sigma=1)
        sigma_y = pm.Exponential("sigma_y", 1) 
        beta_3 = pm.Normal("beta_3", mu=0, sigma=1)
        beta_5 = pm.Normal("beta_5", mu=0, sigma=1)
        beta_6 = pm.Normal("beta_6", mu=0, sigma=1)
        ppID = df_subset['ppID_new'].values

        # 被试随机效应
        if non_centered:
            beta_0_offset = pm.Normal("beta_0_offset", 0, sigma=1, dims="ppID")
            beta_0j = pm.Deterministic("beta_0j", beta_0 + beta_0_offset * beta_0_sigma, dims="ppID")
        else:
            beta_0j = pm.Normal("beta_0j", mu=beta_0, sigma=beta_0_sigma, dims="ppID")

        # 线性关系
        mu = pm.Deterministic("mu", 
                              beta_0j[ppID] +
                              beta_1 * df_subset['Emotion_1'] +
                              beta_3 * df_subset['Time'] * df_subset['Emotion_1'] +
                              beta_5 * df_subset['baseline_mean_RSA'] +
                              beta_6 * df_subset['practice_mean_RSA'],            
                              dims="obs_total")

        # 定义 likelihood
        likelihood = pm.Normal("y_RSA", mu=mu, sigma=sigma_y, observed=df_subset.RSA, dims="obs_total")

        RSA_randomintercept_trace = pm.sample(draws=1000,
                                             tune=300,
                                             chains=4,
                                             discard_tuned_samples=True,
                                             random_seed=84735,
                                             target_accept=0.99)
    
    return RSA_randomintercept_model, RSA_randomintercept_trace

```

```python
RSA_randomintercept_model, RSA_randomintercept_trace = run_RSA_randomintercept_model(non_centered = True)
```

### 模型3：一阶自回归模型  
$$  
 [ \text{RSA}_{\text{randomAR1}} = \beta_0 + \beta_1 \times \text{Time} + \beta_2 \times \text{EmotionCondition} + \beta_3 \times \text{baseline_mean_RSA} + \beta_4 \times \text{practice_mean_RSA} + \beta_5 \times \text{Time} \times \text{EmotionCondition} + \text{random_effects} + \text{correlation_structure} + \epsilon ]  
$$  

level1:被试水平（不同被试的随机截距）  

level2:时间水平（同一被试的不同时间点1-9的时间点，由于每一时间点的数据都和上一个数据相关，因此使用一阶自回归模型，用来表示当前时刻的观测值与上一个时刻的观测值之间存在线性关系）  



```python
def run_RSA_randomAR1_model(non_centered=False):
    # 定义数据坐标，包括被试和观测索引

    coords = {"ppID": df_subset['ppID_new'].values,  
              "obs_total": df_subset['obs_total'].values}

    with pm.Model(coords=coords) as RSA_randomAR1_model:
        # 定义全局参数
        beta_0 = pm.Normal("beta_0", mu=5, sigma=1)
        beta_0_sigma = pm.Exponential("beta_0_sigma", 1)
        beta_time = pm.Normal("beta_time", mu=0, sigma=5)
        beta_emotion_2 = pm.Normal("beta_emotion_2", mu=0, sigma=5)
        beta_time_emotion_2 = pm.Normal("beta_time_emotion_2", mu=0, sigma=5)
        baseline_mean_RSA = pm.Normal("baseline_mean_RSA", mu=0, sigma=5)
        practice_mean_RSA = pm.Normal("practice_mean_RSA", mu=0, sigma=5)
        ppID = df_subset['ppID_new'].values
        
        # 被试随机效应
        if non_centered:
            beta_0_offset = pm.Normal("beta_0_offset", 0, sigma=1, dims="ppID")
            beta_0j = pm.Deterministic("beta_0j", beta_0 + beta_0_offset * beta_0_sigma, dims="ppID")
        else:
            beta_0j = pm.Normal("beta_0j", mu=beta_0, sigma=beta_0_sigma, dims="ppID")

        # 定义时间效应的随机效应（一阶自回归）
        rho = pm.Normal("rho", mu=0, sigma=1, dims="ppID")
        sigma_time = pm.Exponential("sigma_time", 1, dims="ppID")
        time_effect = rho[ppID] * df_subset["Time"].diff().fillna(0)  # 时间差分


        # 线性关系
        mu = pm.Deterministic(
            "mu", 
            beta_0j[ppID] + 
            beta_time * df_subset["Time"] +
            beta_emotion_2 * df_subset["Emotion_1"] + 
            beta_time_emotion_2 * df_subset["Time"] * df_subset["Emotion_1"] +
            baseline_mean_RSA * df_subset["baseline_mean_RSA"] +
            practice_mean_RSA * df_subset["practice_mean_RSA"] +
            time_effect,
            dims="obs_total"
        )

        # 定义 likelihood
        likelihood = pm.Normal("y_RSA", mu=mu, sigma=beta_0_sigma, observed=df_subset["RSA"], dims="obs_total")


        # 进行采样
        RSA_randomAR1_trace = pm.sample(draws=1000, 
                                        tune=300, 
                                        chains=4, 
                                        discard_tuned_samples=True, 
                                        random_seed=84735, 
                                        target_accept=0.99)

    return RSA_randomAR1_model, RSA_randomAR1_trace
```

```python
RSA_randomAR1_model, RSA_randomAR1_trace = run_RSA_randomAR1_model(non_centered = True)
```

```python

# 定义计算 MAE 函数
from statistics import median
def MAE(model_ppc):
    # 计算每个X取值下对应的后验预测模型的均值
    pre_y = model_ppc.posterior_predictive["y_RSA"].stack(sample=("chain", "draw"))
    pre_y_mean = pre_y.mean(axis=1).values

    # 提取观测值Y，提取对应Y值下的后验预测模型的均值
    MAE = pd.DataFrame({
        "RSA_ppc_mean": pre_y_mean,
        "RSA_original": df_subset.RSA
    })

    # 计算预测误差
    MAE["pre_error"] = abs(MAE["RSA_original"] -\
                            MAE["RSA_ppc_mean"])

    # 最后，计算预测误差的中位数
    MAE = median(MAE.pre_error)
    return MAE
```

```python
# 定义计算超出HDI的数量
def counter_outlier(model_ppc, hdi_prob=0.95):
    # 将az.summary生成的结果存到hdi_multi这个变量中，该变量为数据框
    hdi = az.summary(model_ppc, kind="stats", hdi_prob=hdi_prob)
    lower = hdi.iloc[:,2].values
    upper = hdi.iloc[:,3].values

    # 将原数据中的RSA数据合并，便于后续进行判断
    y_obs = model_ppc.observed_data["y_RSA"].values

    # 判断原数据中的RSA分数是否在后验预测的95%可信区间内，并计数
    hdi["verify"] = (y_obs <= lower) | (y_obs >= upper)
    hdi["y_obs"] = y_obs
    hdi_num = sum(hdi["verify"])

    return hdi_num
```

```python
#比较MAE
RSA_fixedintercept_model_MAE = MAE(RSA_fixedintercept_model_ppc)
print(f"固定截距模型MAE: {RSA_fixedintercept_model_MAE:.2f}")

RSA_randomintercept_model_MAE = MAE(RSA_randomintercept_model_ppc)
print(f"随机截距模型MAE: {RSA_randomintercept_model_MAE:.2f}")

RSA_randomAR1_model_MAE = MAE(RSA_randomAR1_model_ppc)
print(f"一阶自回归模型MAE: {RSA_randomAR1_model_MAE:.2f}")
```

### MAE结果解读  
MAE（平均绝对误差）数值越小表示模型的预测误差越小，模型的性能越好。  

固定截距模型的 MAE 是 0.28，随机截距模型的 MAE 是 0.21，一阶自回归模型的 MAE 是 0.17。根据这些数值来看，一阶自回归模型的 MAE 最小，因此在这个比较中，一阶自回归模型的性能最好。

```python
#比较超出HDI的数量
RSA_fixedintercept_model_outliers = counter_outlier(RSA_fixedintercept_model_ppc)
print(f"固定截距模型超出95%hdi: {RSA_fixedintercept_model_outliers:.2f}")

RSA_randomintercept_model_outliers = counter_outlier(RSA_randomintercept_model_ppc)
print(f"随机截距模型超出95%hdi: {RSA_randomintercept_model_outliers:.2f}")

RSA_randomAR1_model_outliers = counter_outlier(RSA_randomAR1_model_ppc)
print(f"一阶自回归模型超出95%hdi: {RSA_randomAR1_model_outliers:.2f}")
```

### HDI结果解读  
95% HDI（高密度区间，High Density Interval）表示对参数的不确定性范围。当HDIs超出1.0时，意味着这些模型的某些参数在95%的置信水平下超出了较宽的不确定性范围。  

固定截距模型的某些参数超出了95% HDI，随机截距模型的某些参数超出了3.00个单位，而一阶自回归模型的某些参数超出了1.00个单位。  

这意味着在这些模型中，一些参数可能存在较大的不确定性，因为它们的后验分布（posterior distribution）在95%的情况下覆盖了较宽的范围。
