### 从计数到概率

一般在统计学中，我们更多使用概率而不是计数。

比如，当抽到“蓝-白-蓝”的石头后，我们可以推测袋子里鹅卵石为一蓝三白的概率为：

#### 几个重要的概念：
* 猜测同的蓝色石头的比例 $p$: 模型参数(parameter);
* 给定$p$值时，得到某种数据的方式: 似然 (likelihood); 
* prior plausibility -->  prior probability 先验概率 (prior)
* 在得到数据后，更新后的$p$：后验概率 (posterior)

## 模型思维：理论的“小世界”与现实的“大世界”

当哥伦布在1492年寻找新大陆时，他坚信地球是个球体。为了船员在寻找新大陆时可以拥有足够的食物，哥伦布根据3万公里的地球周长计算出来航行所需的物资。

虽然地球的实际周长为4万公里，但幸运的事一路上哥伦布的船队遇到了很多岛屿可供他们提供补给，哥伦布的船队也顺利到达了美洲。


![Image Name](https://cdn.kesci.com/upload/image/rhqayo7s64.png?imageView2/0/w/720/h/720)

在哥伦布的眼中的小地球（3万公里）和现实中的大地球（4万公里）就像是统计中的模型和真实情况。


### 统计学中的*小世界*与*大世界*：

小世界：所有的可能性都是事先设定的

大世界：大世界是部署模型的更广泛的环境（或改为“模型运行的真实环境”）

贝叶斯模型从证据中学习的方式在小世界中可以说是最佳的。

当贝叶斯模型的假设与现实相近时，贝叶斯模型在大世界中也表现良好

#### 一个例子并用Python实现

现在使用我们对p和“plausibility”的定义进行更新：

```python
import numpy as np#导入numy库
```

```python
ways = [0, 3, 8, 9, 0]#生成数据的路径数量，等于表格中的Ways to produce data
```

```python
ways#查看一下ways
```

```python
plaus = ways/np.sum(ways)#将ways除ways总和，就是plausibility
plaus
```

```python
p = [0, 0.25, 0.5, 0.75, 1]#推测中鹅卵石为蓝色的占比
```

for 循环

```python
for i, j in zip(p, ways):#zip() 函数用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的列表。
    print(i*j)#将每个i和j乘起来
```

新建一个list_products列表，将p和ways相乘后加进去

```python
list_products = [];
for i, j in zip(p, ways):
    list_products.append(i*j)
```

```python
list_products
```

新加一个sum_products列表

```python
sum_products = np.sum(list_products)
```

观测到新数据后第二种推测发生的概率为：

```python
(3*0.25)/sum_products
```
