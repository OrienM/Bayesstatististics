```python
import numpy as np
import pandas as pd

np.random.seed(123)
```

## STEP1：生成被试编号  
1. 生成100个被试编号，格式为"sub_xx"，如sub_01，sub_02...sub_100  
2. 从其中选取编号为奇数的被试，如sub_01, sub_03,...sub_99  
3. 将结果存为sub_array  

    注：可参考教程中的代码

```python
sub_array = [f"sub{i:02}" for i in range(1,101)]
```

```python
sub_array = sub_array[::2]
```

## STEP2：生成不同条件下的反应时  
1. 生成50个条件A下的反应时，条件A下的反应时服从均值为350，标准差为25的正态分布，变量名存为“rt_A”  
2. 生成50个条件B下的反应时，条件B下的反应时服从均值为450，标准差为35的正态分布，变量名存为“rt_B”  
3. 计算条件A与条件B反应时的差值(条件B减去条件A)，变量名存为“rt_diff”  
* 注：将所有生成的反应时存为整数

```python
rt_A = np.random.normal(loc = 350,
                 scale = 25,
                 size = 50)
rt_A = rt_A.astype(int)
```

```python
rt_B = np.random.normal(loc = 450,
                 scale = 35,
                 size = 50)
rt_B = rt_B.astype(int)
```

```python
rt_diff = rt_B - rt_A
rt_diff = rt_diff.astype(int)
```

## STEP3：转为数据框  
1. 将被试编号、条件A、B的反应时，与二者差值存在同一个数据框中  
2. 列名分别为：sub_ID，rt_A，rt_B，rt_diff  
3. 数据框命名为df_rt  
4. 选取rt_diff 列的前10个值，存入变量a1中

```python
df_rt = pd.DataFrame({'sub_ID': sub_array,
                      'rt_A': rt_A, 
                      'rt_B': rt_B, 
                      'rt_diff': rt_diff})
```

```python
df_rt
```

```python
a1 = df_rt['rt_diff'][0:10]
```
