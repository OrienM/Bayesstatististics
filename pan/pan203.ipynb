# 频率学派与贝叶斯学派的对比

## 频率学派如何看待这个世界？


在对比频率学派与贝叶斯学派的差异之前，让我们首先回顾一下频率学派是如何看待这个世界的。

```python
Image("./figure/pan20301.png", width='600px') # 图片来源 https://www.bilibili.com/video/BV1NG411b7qD
```

我们假设存在两个城市，其中A城市的平均工资为$6000$元每月，B城市的平均工资为$6500$元每月。

然后，由于两个城市的人口太多，通过两个城市中所有人的工资去计算平均工作是*费时费力的*。其次，对于‘地球中大陆的占比’的问题，记录所有的陆地和海洋是不太可能的。

因此，统计学家选择 **抽样** 的方式通过 **样本** 来估计两个城市 **总体** 的均值。

值得注意的是：
1. 这个例子中两个总体均值的差异(工资差异)是**固定的**，即世界的真相是固定的。然而对于更复杂的问题，比如两个人性格的差异，这类世界现象是难以确定其背后的差异是否固定。
2. 通过抽样的方式去估计总体会存在**噪音**或者*偏差*。但对于经过传统统计训练的学习者来说，这是一种直接思维，但我们是否想过有另一种方式去看待噪音？
3. 频率学派认为**概率**是噪音的体现。结合前两点，两个城市的差异是固定的，而抽样带来了误差，因此概率代表的是样本能代表总体的概率。
4. 噪音受到抽样**样本大小**和工资方差**变异**的影响。另一个常见的问题是抽样样本的概率分布与样本大小和总体方差有关，如下图。

最后，频率学派如何**推断**出两个总体之间的差异？
- 由于两个总体间的差异是固定的，因此，判断差异是否显著与噪声大小有关，即信噪比。
- 频率学派通过零假设的显著性检验(Null hypothesis significant test, NHST)来判断这个显著性。即通过拒绝极端值的方式避免噪音的影响。在数学上通过计算置信区间(confidence interval)和$p$值来帮助推断过程。

## 贝叶斯学派如何看待这个世界？

贝叶斯学派的出现与兴趣在于频率学派所存在的问题。
主要体现在2个方面：
1. 世界的真相(差异)可能不是固定的。比如对于复杂世界现象背后的机制是难以确定的。
2. 通过抽样样本去替代总体容易出现偏差。比如由于抽到极端值从而错误的估计总体。此外，相关的问题还有typeⅠ错误，统计检验力，p-hacking等。

贝叶斯学派提供了另一种角度去解释上述的问题。

- 首先，贝叶斯学派认为世界的真相不是固定不变的，并且他们通过概率去描述这种变化。比如，两个城市平均工资的差异可能受到各种因素的影响，因此这个差异不是固定的。而两个城市平均工资的差异或可能性，比如这个差异为$500$的概率为$80$%，代表了个体对这个差异的信念**(belief)**，比如有$80$%的把握相信这个差异为$500$。

- 其次，贝叶斯学派并不是通过NHST来推测这个差异。而是通过贝叶斯公式，即贝叶斯定理。需要注意的是，虽然贝叶斯学派认为世界的真相并不是固定的，但总世界中所观测的现象(抽样样本得到的数据)是固定的。因此，对于世界真相的信念可以根据数据进行更新。

## 两个学派的差异对比

|                     | 频率学派   | 贝叶斯学派   |
| ------------------- | ---------- | ------------ |
| 世界真相 (总体差异) | 固定       | 变化         |
| 概率                | 抽样的噪音 | 信念         |
| 推断过程            | NHST       | 贝叶斯定理   |
| 数据                | 存在噪音   | 固定         |
| 推断可更新性        | 否         | 是           |
| 主观性              | 前提假设   | 通过先验设定 |

### 贝叶斯的主观性


因为贝叶斯分析方法中的先验概率依赖于主观设定，因此常被频率学派诟病‘不够客观’。

实际上，任何统计分析方法都不可能完全客观，因此主观性是一个相对概念。

贝叶斯学派的主观性通过先验的设定来体现。而频率学派的主观性暗含在各种**前提假设**中，比如方差分析中的方差齐性和正态性，这种看似‘客观的’假设，一方面难以满足，一方面也是一种主观的设定。

更为宏观的来说，样本的抽取，数据清理方式的选择，分析方法的选择，p值的设定，这些都存在主观性。因此，频率学派并没有想象的那么‘客观’。

一种易于理解的方式是将贝叶斯看作频率学派的扩展，即在通过数据对世界真相进行推断的基础上整合了个体先前的主观经验。


### 两个学派对于重复获取数据和数据分布的看法

频率学派认为，在一个总体中重复获取数据会形成**抽样分布**，即只要无限(long-run)的进行抽样，样本分布的均值就会接近总体均值。通过单次获取的样本数据对抽样分布估计时可以计算置信区间以及相应的p值。

因此，置信区间代表的是总体的真实值率落在测量结果的周围(置信区间)的概率。然而在实际操作中，我们往往只会收集一次数据，并不会反复的进行抽样。因此，如果抽样结果是一个极端值，那么相应的估计都是有错误的。

贝叶斯学派中容易与置信区间(confidence interval)相混淆的概念是可信区间(credible interval)，后者指真实值出现的概率，即对于真实值在该区间某个位置的信念。

虽然两者在概念上截然相反，前者本质是抽样分布后者是真实值分布，但实际上可信区间就是置信区间和可信区间的一种结合。

但更为重要的是，正是可信区间的这种结合方式，使得置信区间在因极端值产生错误估计时可以通过先验信息进行一定的修正；此外，在连续反复的获取数据时，可信区间可以根据新的数据进行**更新**(如下图)，这是频率学派所不具备的。

但两个学派各有优势和缺陷。贝叶斯推断的最大问题就是，如果先验信息设置本身存在偏差，那么即使数据不受极端值的影响，所估计的结果也会存在偏差。

### 关于NHST的问题

如前文中所说，NHST通过拒绝极端值进行统计推断。然而，NHST却无法对于零假设(null hypothesis)进行支持，即如果两个总体没有显著差异，他们的相似程度有多少？

而贝叶斯推断本身就是对两个总体差异的可能性进行估计，因此可以提供对于零假设支持的证据。

此外，NHST一次性只能对比两个总体的假设进行比较，对于多个总体，此时往往需要进行两两比较，并且需要通过Bonferroni等方法校正。
而贝叶斯推断可以同时对多个总体或者假设给出不同的概率估计，即通过条件概率的方式将各种假设作为条件再对参数进行估计。
